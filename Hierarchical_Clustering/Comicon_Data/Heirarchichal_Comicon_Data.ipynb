#                            Hierarchical Clustering - Comicon Dataset

### I have used the following tools and concepts to build a model for the Comicon Dataset:
* Pandas
* Numpy
* Matplotlib 
* Scipy
* Hierarchical-Clustering
* Dendrogram
* Seaborn
* Ward, Single, Complete methods for the linkage() function

```python
# Imorting libraries
import pandas as pd
from scipy.cluster.vq import whiten, kmeans, vq
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import linkage, fcluster
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
comic_con = pd.read_csv("comic-con.csv")
comic_con.head()
```

```python
# Scaling the data using whiten method (dividing the variables with thier std)
comic_con['x_scaled'] = whiten(comic_con['x_coordinate'])
comic_con['y_scaled'] = whiten(comic_con['y_coordinate'])
```

```python
# Exploring the dataset after adding scaled data
comic_con.head()
```

```python
# Using the linkage() function with the single method
distance_matrix = linkage(comic_con[['x_scaled', 'y_scaled']], method = 'single', metric = 'euclidean')

# Creating a dendrogram
dn = dendrogram(distance_matrix)

# Displaying the dendogram
plt.show()
```

```python
# Assigning cluster labels
comic_con['cluster_labels'] = fcluster(distance_matrix, 2, criterion='maxclust')
comic_con.cluster_labels

# Plotting clusters using sns
sns.scatterplot(x='x_scaled', y='y_scaled',
                hue='cluster_labels', data = comic_con)
plt.show()
```

```python
# Using the linkage() function with the complete method
distance_matrix = linkage(comic_con[['x_scaled', 'y_scaled']], method = 'complete', metric = 'euclidean')

# Creating a dendrogram
dn = dendrogram(distance_matrix)

# Displaying the dendogram
plt.show()
```

```python
# Assigning cluster labels
comic_con['cluster_labels'] = fcluster(distance_matrix, 2, criterion='maxclust')
comic_con.cluster_labels

# Plotting clusters
sns.scatterplot(x='x_scaled', y='y_scaled',
                hue='cluster_labels', data = comic_con)
plt.show()
```

```python
# Using the linkage() function with the ward method
distance_matrix = linkage(comic_con[['x_scaled', 'y_scaled']], method = 'ward', metric = 'euclidean')

# Creating a dendrogram
dn = dendrogram(distance_matrix)

# Displaying the dendogram
plt.show()
```

```python
# Assigning cluster labels
comic_con['cluster_labels'] = fcluster(distance_matrix, 2, criterion='maxclust')
comic_con.cluster_labels

# Plotting clusters
sns.scatterplot(x='x_scaled', y='y_scaled',
                hue='cluster_labels', data = comic_con)
plt.show()
```
